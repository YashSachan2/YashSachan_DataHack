```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
# Identify numerical and categorical columns
numerical_features = df.select_dtypes(include=['float64']).columns
categorical_features = df.select_dtypes(include=['object']).columns

# Define preprocessing steps for numerical features
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  # or 'median'
    ('scaler', StandardScaler())
])

# Define preprocessing steps for categorical features
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# Apply the preprocessing pipeline
processed_df = preprocessor.fit_transform(df)
processed_df1=preprocessor.fit_transform(df)

# If you need to convert the resulting numpy array back to a DataFrame
# Note: OneHotEncoder creates multiple columns, so we must create appropriate column names
num_columns = preprocessor.transformers_[0][2]
cat_columns = preprocessor.transformers_[1][1]['onehot'].get_feature_names_out(categorical_features)
all_columns = num_columns.tolist() + cat_columns.tolist()

processed_df = pd.DataFrame(processed_df, columns=all_columns)
processed_df1 = pd.DataFrame(processed_df1, columns=all_columns)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def calculateXValScore(model):
    # Use accuracy as the scoring metric
    scorer = 'accuracy'

    # Perform cross-validation and get mean score
    xValScore = cross_val_score(model, processed_df, df1['xyz_vaccine'], cv=10, scoring=scorer).mean()
    xValScore = round(xValScore * 100, 2)

    return xValScore
```

```python
def calculateXValScore1(model):
    # Use accuracy as the scoring metric
    scorer = 'accuracy'

    # Perform cross-validation and get mean score
    xValScore = cross_val_score(model, processed_df, df1['seasonal_vaccine'], cv=10, scoring=scorer).mean()
    xValScore = round(xValScore * 100, 2)

    return xValScore
```

```python
classifiers = [
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),
    ('Logistic Regression', LogisticRegression(max_iter=1000, random_state=42)),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('Support Vector Classifier', SVC(random_state=42)),
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('Naive Bayes', GaussianNB()),
    ('XGBoost', XGBClassifier(random_state=42)),
    ('LightGBM', LGBMClassifier(random_state=42))
]

for name, model in classifiers:
    xValScore = calculateXValScore(model)
    print(f"{name}: {xValScore}%")
```
